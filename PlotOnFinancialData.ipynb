```python
#Data of revenue and expenses for each month in a year
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
```

```python
months = ['january','february','march','april','may','june','july','august','september','october','november','december']
```

```python
#profit for each month
import numpy as np
profit = np.array(revenue)-np.array(expenses)
```

```python
profit
```

```python
#profit after tax of 30%
profit_after_tax = profit - 0.3*profit
```

```python
profit_after_tax
```

```python
plt.plot(revenue, color="black", ls="--", marker="o", ms=10) 
#ls means line of sight and ms means marker size
plt.xticks(list(range(12)), months)
plt.show()
```

```python
plt.plot(revenue, color="black", ls="--", marker="o", ms=6) 
plt.plot(expenses, color="red", ls="--", marker="+", ms=6)
plt.plot(profit, color="blue", ls="--", marker="s", ms=6)
plt.plot(profit_after_tax, color="green", ls="--", marker="^", ms=6)
#ls means line of sight and ms means marker size
plt.xticks(list(range(12)), months, rotation="vertical")
plt.show()
```

```python
plt.plot(revenue, color="black", ls="--", marker="o", ms=6, label="revenue") 
plt.plot(expenses, color="red", ls="--", marker="+", ms=6, label="expenses")
plt.plot(profit, color="blue", ls="--", marker="s", ms=6, label="profit")
plt.plot(profit_after_tax, color="green", ls="--", marker="^", ms=6, label="profit_after_tax")
#ls means line of sight and ms means marker size
plt.legend() #to show the labels
plt.xticks(list(range(12)), months, rotation="vertical")
plt.show()
```

```python
plt.plot(revenue, color="black", ls="--", marker="o", ms=6, label="revenue") 
plt.plot(expenses, color="red", ls="--", marker="+", ms=6, label="expenses")
plt.plot(profit, color="blue", ls="--", marker="s", ms=6, label="profit")
plt.plot(profit_after_tax, color="green", ls="--", marker="^", ms=6, label="profit_after_tax")
#ls means line of sight and ms means marker size
plt.legend(loc = 'upper left', bbox_to_anchor=(1,1)) #to show the labels at proper location
plt.xticks(list(range(12)), months, rotation="vertical")
plt.show()
```

```python
# club it all together
finance_stat = np.array([revenue, expenses, profit, profit_after_tax])
```

```python
print(finance_stat)
```

```python
featurelist = ["revenue", "expenses", "profit", "profit_after_tax"]
fdict = {"revenue":0, "expenses":1, "profit":2, "profit_after_tax":2}
```

```python
def myplot(data,featurelist):
    for f in featurelist:
        plt.plot(data[fdict[f]],color="blue", ls = "--", marker="o", ms=6, label=f)
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(12)), months, rotation='vertical')
    plt.show()
```

```python
myplot(finance_stat,["revenue", "expenses", "profit"])
#here due to same marker and colour the lines are not much distinguishable
```

```python
#to make the function more advance for distinguishing colour and marker
def myplot(data,featurelist):
    cdict = {"revenue":'black', "expenses":'red', "profit":'blue', "profit_after_tax":'green'}
    mdict = {"revenue":'o', "expenses":'+', "profit":'s', "profit_after_tax":'^'}
    for f in featurelist:
        plt.plot(data[fdict[f]],color = cdict[f], ls = "--", marker="o", ms=6, label=f)
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(12)), months, rotation='vertical')
    plt.show()
```

```python
myplot(finance_stat,["revenue", "expenses", "profit"])
```
