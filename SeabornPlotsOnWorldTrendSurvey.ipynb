```python
#import all dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#ignoring warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
#import data
df = pd.read_csv("World_trend_survey.csv")
```

```python
df.head()
```

```python
#distribution
vis1 = sns.distplot(df["Internet users"],bins=10)
```

```python
vis2 = sns.distplot(df["LifeExp2013"],bins=10)
```

```python
f, ax = plt.subplots(2,2,figsize=(8,4))
vis1 = sns.distplot(df["Internet users"],bins=10, ax= ax[0][0])
vis2 = sns.distplot(df["Birth rate"],bins=10, ax=ax[0][1])
vis3 = sns.distplot(df["LifeExp1960"],bins=10, ax=ax[1][0])
vis4 = sns.distplot(df["LifeExp2013"],bins=10, ax=ax[1][1])
f.savefig('subplot.png')
```

```python
#box plot
vis2 = sns.boxplot(data = df, x = "Income Group", y = "Birth rate")
fig = vis2.get_figure()
fig.savefig("sb6.png")
```

```python
vis3 = sns.boxplot(data = df, x = "Income Group", y = "Internet users")
```

```python
vis4 = sns.boxplot(data = df, x = "Country Region", y = "Internet users")
```

```python
vis5 = sns.boxplot(data = df, x = "Country Region", y = "Birth rate")
```

```python
vis6 = sns.jointplot(data = df, x = "Internet users", y = "Birth rate", kind='kde')
#pearson coefficient -0.82 means its a linear relationship with negative slope
#p value<0.001 means the infos statistically highly significant
#https://www.statsdirect.com/help/basics/p_values.htm
vis6.savefig("sb1.png")
```

```python
vis6 = sns.lmplot(data = df, x = "Internet users", y = "Birth rate",fit_reg=False)
#fig = vis6.get_figure()
vis6.savefig("sb2.png")
```

```python
vis7 = sns.lmplot(data = df, x = "Internet users", y = "Birth rate",fit_reg=False, hue = "Income Group",\
                 size = 6, aspect=1.5, scatter_kws = {'s':200}, )
vis7.savefig("sb3.png")
```

```python
vis8= sns.lmplot(data = df, x = "Internet users", y = "Birth rate",fit_reg=False,\
                  hue = "Country Region",\
                 size = 6, aspect=1.5, scatter_kws = {'s':200}, )
vis8.savefig("sb5.png")
```

```python
vis7 = sns.lmplot(data = df, x = "LifeExp1960", y = "LifeExp2013",fit_reg=False, hue = "Country Region",\
                 size = 6, aspect=1.5, scatter_kws = {'s':200}, )
```

```python
ax = sns.swarmplot(x="LifeExp1960", y="Country Region", hue="Income Group", data=df)
ax.legend_.remove()
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```

```python
ax = sns.swarmplot(x="LifeExp2013", y="Country Region", hue="Income Group", data=df)
ax.legend_.remove()
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```
